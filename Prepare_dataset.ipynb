```python
import os
from PIL import Image

def undo_cropping(image_parts, output_path):
    # This function restores the original image by combining the cropped parts.

    if len(image_parts) == 0:
        print("No image parts found.")
        return

    # Determine the size of the restored image
    part_width, part_height = image_parts[0].size
    total_width = part_width * 2
    total_height = part_height * 2

    # Create a new blank image to store the restored image
    restored_image = Image.new('RGB', (total_width, total_height))

    # Paste each image part onto the restored image
    for i in range(2):
        for j in range(2):
            part_image = image_parts[i * 2 + j]
            x = j * part_width
            y = i * part_height
            restored_image.paste(part_image, (x, y))

    # Save the restored image
    restored_image.save(output_path)
    print("Image restoration complete.")


# Specify the folder path containing the cropped image parts
folder_path = 'prueba'

# Specify the output path for the restored image
output_path = 'reconstructed/restored_image.jpg'

# Collect the cropped image parts
image_parts = []
for file_name in os.listdir(folder_path):
    if file_name.endswith(".jpg"):
        file_path = os.path.join(folder_path, file_name)
        image_part = Image.open(file_path)
        image_parts.append(image_part)

# Undo the cropping and restore the original image
undo_cropping(image_parts, output_path)

```

## Funcion para escribir los nombres de los archivos a un csv y poder tratarlos en excel

## Funcion para renombrar los datos. Usa el nombre de la carpeta contenedora, que es la categoria + el id, que es lo que est√° entre el file extension y el 20x. Es la que he usado


```python
import os
import re

def rename_files(root_dir):
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            original_path = os.path.join(root, file)
            folder_name = os.path.basename(root)
            magnification = get_magnification(file)
            file_id = get_id(file)
            new_name = f"{folder_name}_{magnification}_{file_id}.jpg"  # Adjust the file extension as needed

            new_path = os.path.join(root, new_name)
            os.rename(original_path, new_path)
            print(f"Renamed file: {file} to {new_name}")

def get_magnification(file_name):
    if '20x' in file_name:
        return '20x'
    elif '40x' in file_name:
        return '40x'
    else:
        return ''  # Return an empty string if the magnification is not found

def get_id(file_name):
    id_match = re.search(r"\d+(?=\.)", file_name)  # Matches one or more digits before the file extension
    if id_match:
        return id_match.group()
    else:
        return ''  # Return an empty string if the ID is not found

# Provide the root directory path
root_directory = 'julio'

# Call the function to rename the files
rename_files(root_directory)
```
