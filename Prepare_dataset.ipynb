```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.utils import load_img

import os

import seaborn as sns
from matplotlib import pyplot as plt

from fnmatch import fnmatch
```

```python
# Funcion para cortar en trozos

def crop_directory_subdirectory(path, n_horizontal, n_vertical):

    #This function takes a "root" directory and crops all the images in all the subdirectories, crops it in n_pieces, and then removes the ooriginal image

    for subdir, dirs, files in os.walk(path):
        for file in files:       
            
            f = os.path.join(subdir, file)
            im = tf.keras.utils.load_img(f)
            
            imgwidth, imgheight = im.size
            

            trozos_h = int(imgwidth/n_horizontal)
            trozos_v = int(imgheight/n_vertical)
            
            

            for i in range(n_horizontal):
                for j in range(n_vertical):
                    
                    print(f)

                    # print("i = ", i)
                    # print("j = ", j)
                    #print( trozos_v*j ,trozos_h*i ,trozos_v*j + trozos_v,trozos_h*i+ trozos_h)        
                        
                    trozo = im.crop((trozos_h*j ,trozos_v*i ,trozos_h*j + trozos_h,trozos_v*i+ trozos_v))    
                    

                    trozo.save(f + str(i) + "_" + str(j)+ ".jpg")


            os.remove(f)

            
```
